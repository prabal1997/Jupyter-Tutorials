```python
#EXERCISE 1 & 2
import numpy as np

def my_factorial(n):
    product = 1;
    for x in range(1, n+1):
        product *= x;
    return product;

def binomial_rv(n, p):
    success_count = 0; #counts the number of successful tosses
    for x in range(n):
        if (np.random.uniform() < p):
            success_count += 1;
            
    return success_count;

print(binomial_rv(10, 0.5));
```

```python
#EXERCISE 3
import numpy as np

def give_pi_approx(trial_count):
    total_count_in_circle = 0;
    for element in range(0, trial_count):
        x_axis, y_axis = [np.random.uniform(), np.random.uniform()];
        radius = np.sqrt(x_axis**2 + y_axis**2);
        if (radius < 1):
            total_count_in_circle += 1;
        
    return 4*total_count_in_circle/trial_count;

print(give_pi_approx(5000));
```

```python
#EXERCISE 6 (Exercise 5 is a SUBSET of this)
import numpy as np
import matplotlib.pyplot as plt

#assign values to constants
T_val = 200;
alpha_array = [0, 0.8, 0.98];
col_array = ['b-', 'g-', 'r-'];

#generate list of random, normally distributed numbers for 'shock' use
rand_array = np.random.randn(T_val);

for num in range(0, 3):
    #create lists for storing final values
    x_array = np.zeros(T_val+1);
    for idx in range(1, T_val+1):
        x_array[idx] = alpha_array[num]*x_array[idx-1] + rand_array[idx-1];

    plt.plot(x_array, col_array[num], label=("alpha = " + str(alpha_array[num])));

plt.legend();
plt.show();
```
